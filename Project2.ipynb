```python
sns.factorplot(data=df_clean, x='ASSESSMENT_NBHD',y='PRICE')

#factorplot(data=DataFrame, x=str, y=str, hue=str)

#rotate x-axis labels on ALL charts
#g.set_xticklabels(rotation=-45)
```

```python
plt.figure(figsize=(13,7))

sns.swarmplot(x='ASSESSMENT_NBHD', y='PRICE',
              data=df_clean)
             
#adjust the y-axis
plt.ylim(0,260)

#put the legend on the right
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
#change size of plot
plt.subplots(figsize=(18,7))

#violin plot shows the distribution of prices per neighborhood
sns.violinplot(x='ASSESSMENT_NBHD', y='PRICE', data=df_clean)
plt.xticks(rotation=-45)
plt.title('Neigborhood vs. Assessments')
```
